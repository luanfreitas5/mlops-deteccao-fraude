Exploração do dataset Credit Card Fraud (ULB). Foco no desbalanceamento extremo,

```python
import matplotlib.pyplot as plt
import numpy as np
import polars as pl
import seaborn as sns

from src.config.paths import get_paths
from src.visualization.theme import apply_theme

RANDOM_SEED = 42
rng = np.random.default_rng(seed=RANDOM_SEED)
rng.normal()
apply_theme()
```

```python
df = pl.read_csv(get_paths().data_raw)
print(df.shape)
df.head()
```

## Desbalanceamento de classes

```python
contagem = df.group_by("Class").len().sort("Class")
taxa_fraude = df["Class"].mean()
print(f"Taxa de fraude: {taxa_fraude:.4%}")
contagem
```

## Distribuição do valor da transação por classe

```python
pdf = df.select(["Amount", "Class"]).to_pandas()
fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(data=pdf, x="Amount", hue="Class", bins=50, log_scale=(False, True), ax=ax)
ax.set_title("Distribuição do valor por classe (escala log no eixo Y)")
ax.set_xlabel("Valor da transação (Amount)")
ax.set_ylabel("Frequência (log)")
plt.tight_layout()
```
